```python
average_age = df['age'].mean()
print("Average Age :",average_age*100)
```

```python
sns.histplot(df['income'])
plt.xlabel('Income')
plt.ylabel('Count')
plt.title('Income Distribution')
plt.show()
```

```python
illness_counts = df['illness'].value_counts()
sns.barplot(x=illness_counts.index, y=illness_counts.values)
plt.xlabel('Illness')
plt.ylabel('Count')
plt.title('Number of Visits by Illness Status')
plt.show()
```

```python
sns.scatterplot(x=df['age'], y=df['income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age versus Income')
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
avg_visits_by_gender = df.groupby('gender')['visits'].mean()
print(avg_visits_by_gender)
```

```python
max_income_by_illness = df.groupby('illness')['income'].max()
print(max_income_by_illness)
```

```python
sns.boxplot(x=df['gender'], y=df['income'])
plt.xlabel('Gender')
plt.ylabel('Income')
plt.title('Income by Gender')
plt.show()
```

```python
bins = [0, 1, 2, 3, 4, 5]  # Define age groups
age_groups = pd.cut(df['age'], bins)
age_group_counts = age_groups.value_counts().sort_index()
sns.barplot(x=age_group_counts.index.astype(str), y=age_group_counts.values)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Number of Visits by Age Group')
plt.show()
```
